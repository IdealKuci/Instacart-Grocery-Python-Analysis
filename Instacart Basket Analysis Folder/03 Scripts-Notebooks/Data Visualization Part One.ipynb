```python
# Mixed d.types check: First name contains mixed d.types
for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

```python
# Converting the mixed first_name d.type to a string 
customers['first_name'] = customers['first_name'].astype('str')
```

```python
# Checking for mix d.types after converting the mixed first_name d.type to a string
# No more mixed types found 
for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```
